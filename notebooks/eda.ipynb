Patrones de faltantes en el dataset

El an√°lisis exploratorio de los datos mostr√≥ que las variables tienen diferentes proporciones y patrones de faltantes, no siendo un problema meramente aleatorio. 

1. A partir del heatmap de las primeras 1000 filas, se observ√≥ que los faltantes siguen un patr√≥n estructurado. Por ejemplo, variables relacionadas con micronutrientes (hierro, zinc, vitaminas) y alimentaci√≥n (diversidad, lactancia) tienen bloques completos sin datos. Esto indica que los registros con estas mediciones fueron aplicados solo a ciertos subgrupos, es decir, los datos no est√°n ausentes al azar (Missing Not At Random, MNAR).

2. Al analizar los faltantes por grupos de edad:
- La variable `zscore_imc_para_edad` tiene un 100% de faltantes en ni√±os menores de 5 a√±os. Solo se encuentra disponible a partir del grupo de 5 a 10 a√±os, lo que indica que este indicador se calcul√≥ o registr√≥ √∫nicamente para poblaciones mayores.
- En contraste, la variable `hemoglobina_ajustada_g_dl` muestra alrededor de 16% a 18% de faltantes distribuidos homog√©neamente en todos los grupos etarios. Esto sugiere un patr√≥n m√°s cercano a MAR (Missing At Random), probablemente debido a decisiones log√≠sticas de medici√≥n.

3. Al analizar los faltantes por sexo, se observ√≥ que las proporciones son pr√°cticamente iguales en ni√±os y ni√±as. Por ejemplo, `zscore_imc_para_edad` tiene alrededor de 39% a 40% de faltantes en ambos sexos, y `hemoglobina_ajustada_g_dl` alrededor de 22%. Esto indica que el patr√≥n de ausencia no depende del sexo.

Conclusi√≥n pr√°ctica: las variables con m√°s del 50% de datos faltantes no aportan informaci√≥n robusta y deben eliminarse del an√°lisis principal. En cambio, variables con entre 10% y 50% de faltantes son candidatas a imputaci√≥n por mediana (si son num√©ricas) o moda (si son categ√≥ricas) u otras t√©cnicas m√°s avanzadas (lo veremos m√°s adelante). Adem√°s, tal vez para variables como `zscore_imc_para_edad` resulta recomendable restringir el an√°lisis a poblaciones mayores de 5 a√±os, que es donde el dato realmente existe, evitando imputar valores que nunca fueron medidos o buscar otra soluci√≥n.



```python
# Calcular porcentaje de faltantes
faltantes = df.isna().sum().to_frame('n_faltantes')
faltantes['pct_faltantes'] = (faltantes['n_faltantes'] / len(df)) * 100
faltantes.sort_values('pct_faltantes', ascending=False, inplace=True)

# Identificar columnas a eliminar (>50% faltantes)
cols_a_eliminar = faltantes[faltantes['pct_faltantes'] > 50].index.tolist()
print(f"Columnas a eliminar por >50% faltantes ({len(cols_a_eliminar)} columnas):")
print(cols_a_eliminar)

# Eliminar columnas
df.drop(columns=cols_a_eliminar, inplace=True)
print("Shape despu√©s de eliminar columnas:", df.shape)
```

```python
# Ver columnas con faltantes restantes
faltantes_restantes = df.isna().sum().to_frame('n_faltantes')
faltantes_restantes['pct_faltantes'] = (faltantes_restantes['n_faltantes'] / len(df)) * 100

# Filtrar solo las que tienen alg√∫n faltante
faltantes_restantes = faltantes_restantes[faltantes_restantes['n_faltantes'] > 0]
faltantes_restantes = faltantes_restantes.sort_values('pct_faltantes', ascending=False)

print("Columnas que a√∫n tienen faltantes (‚â§50%):")
print(faltantes_restantes)
```

#### 1: Imputar mediana o moda en las variables ‚Äúno cr√≠ticas‚Äù (puede cambiar)

```python
# Definir todas las columnas importantes, incluyendo crecimiento y jer√°rquicas
columnas_importantes = [
    'zscore_imc_para_edad', 'clasificacion_imc',
    'anemia', 'hemoglobina_ajustada_g_dl', 'hemoglobina_g_dl',
    'indice_socioeconomico', 'nivel_socioecon_tercil', 'nivel_socioecon_quintil', 
    'nivel_socioecon_decil', 'estrato_varianza', 'estrato_dise√±o', 
    'estrato_urbanidad', 'estrato_marginalidad', 
    'area_urbana_rural', 'region_geo', 'altitud_localidad'
]

# Identificar columnas con faltantes que no son cr√≠ticas
columnas_faltantes_no_criticas = [
    col for col in df.columns 
    if df[col].isna().sum() > 0 and col not in columnas_importantes
]

# Imputar mediana/moda en esas columnas no cr√≠ticas
for col in columnas_faltantes_no_criticas:
    if df[col].dtype in ['float64', 'int64']:
        mediana = df[col].median()
        df[col].fillna(mediana, inplace=True)
        print(f"Imputada columna num√©rica: {col} con mediana {mediana}")
    else:
        moda = df[col].mode()[0]
        df[col].fillna(moda, inplace=True)
        print(f"Imputada columna categ√≥rica: {col} con moda {moda}")
```

####  Imputaci√≥n avanzada con IterativeImputer (MICE-like) para las variables cr√≠ticas

Imputaci√≥n avanzada con IterativeImputer

Se utiliza IterativeImputer (implementaci√≥n de sklearn inspirada en MICE - Multiple Imputation by Chained Equations) para imputar valores faltantes en las variables cr√≠ticas del modelo jer√°rquico: zscore_imc_para_edad, clasificacion_imc, anemia, hemoglobina, as√≠ como los indicadores socioecon√≥micos y de estrato. Este m√©todo itera ajustando modelos de regresi√≥n para cada variable con faltantes, usando las dem√°s como predictoras, respetando as√≠ la estructura multivariada del dataset y preservando correlaciones naturales entre IMC, anemia, altitud y nivel socioecon√≥mico. Esto permite imputar de forma m√°s coherente que simplemente usar medias, modas o distancias de KNN.


```python
# Aqu√≠ usamos IterativeImputer, una t√©cnica inspirada en MICE (Multiple Imputation by Chained Equations).
# Elegimos esto en lugar de KNN porque:
# - Modela expl√≠citamente dependencias estad√≠sticas entre variables
# - Es robusto ante multicolinealidad y aprovecha la estructura jer√°rquica: ej. hemoglobina ~ IMC + edad + estrato.
# - Es preferido cuando hay m√∫ltiples variables con NA interdependientes (caso t√≠pico en encuestas nutricionales).

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Extraemos solo el subset con las variables importantes
subset_critico = df[columnas_importantes]

# Configuramos IterativeImputer
# - max_iter=10 para que itere refinando imputaciones
# - random_state para reproducibilidad
imputer_mice = IterativeImputer(max_iter=10, random_state=42)

# Ajustamos el imputador y transformamos los datos
subset_critico_imputado = pd.DataFrame(
    imputer_mice.fit_transform(subset_critico),
    columns=columnas_importantes
)

# Sustituimos las columnas originales con el resultado imputado
df[columnas_importantes] = subset_critico_imputado

print("‚úÖ Imputaci√≥n avanzada con IterativeImputer completada en variables cr√≠ticas.")
```

```python
# Verificamos que ya no quedan faltantes en el dataframe completo
faltantes_restantes = df.isna().sum()
faltantes_restantes = faltantes_restantes[faltantes_restantes > 0]

print("\nColumnas con faltantes finales (se supone que deber√≠a estar vac√≠o):")
print(faltantes_restantes)

print("\nShape final del dataframe limpio:", df.shape)
```

- Si una fila tiene NAs en todas las columnas usadas en el imputador, no puede imputar nada, y deja los NAs.
- Por ejemplo, si ese ni√±o no tiene ni IMC, ni hemoglobina, ni socioecon√≥mico, ni nada, IterativeImputer no tiene forma de predecirlo.

- Solo hay 881 filas (‚âà3%) que quedaron con NAs porque est√°n totalmente incompletas en todas esas columnas cr√≠ticas.

- Es normal, y lo correcto es simplemente eliminarlas, porque no aportan nada al modelo (no tienen ninguno de los predictores importantes).



```python
# Eliminar solo las filas que sigan teniendo NAs en esas columnas importantes
df.dropna(subset=columnas_importantes, inplace=True)

print("Nuevo shape tras eliminar filas sin datos cr√≠ticos:", df.shape)
print("Faltantes tras limpieza final (ahora si deber√≠a estar vac√≠o):")
print(df[columnas_importantes].isna().sum())
```
